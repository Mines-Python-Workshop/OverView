## 1.1.3 Importing multiple files simultaneously

```python
import glob
```

```python
all_files = glob.glob('./Data/*')
print(all_files)
```

```python
csv_files = glob.glob('./Data/*.csv')
print(csv_files)
```

```python
csv_dfs = []

for i in range(len(csv_files)):
	df = pd.read_csv(csv_files[i])
	csv_dfs.append(df)
	
for i in range(len(csv_dfs)):
	print(csv_dfs[i])
```
